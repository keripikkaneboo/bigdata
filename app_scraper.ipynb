```python
from google_play_scraper import reviews
from google_play_scraper import Sort
import pandas as pd

# ---------------------------------------
# Parameter
APP_ID   = 'com.gojek.app'   # ID aplikasi
N_DATA   = 50000              # jumlah baris yang ingin diambil
LANG     = 'id'              # bahasa
COUNTRY  = 'id'              # negara
BATCH    = 200               # jumlah maksimum review per request (disarankan <=200)
# ---------------------------------------

all_reviews = []
count = 0

while count < N_DATA:
    # Ambil batch review terbaru
    batch_reviews, _ = reviews(
        APP_ID,
        lang=LANG,
        country=COUNTRY,
        sort=Sort.NEWEST,
        count=BATCH,
        continuation_token=None  # token otomatis diganti setiap iterasi
    )
    all_reviews.extend(batch_reviews)
    count = len(all_reviews)

    # Jika sudah cukup, hentikan
    if len(batch_reviews) < BATCH:
        break  # Tidak ada review lagi

# Potong agar pas sesuai N_DATA
all_reviews = all_reviews[:N_DATA]

# Simpan ke DataFrame
app_reviews_df = pd.DataFrame(all_reviews)
app_reviews_df.to_csv('review_gojek.csv', index=False, header=True)

print(f"Total review tersimpan: {len(app_reviews_df)}")
display(app_reviews_df.head())
```
