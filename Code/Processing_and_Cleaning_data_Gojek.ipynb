```python
import pandas as pd

path_stream = "/content/review_gojek.csv"
path_batch  = "/content/gojek.csv"

df_stream = pd.read_csv(path_stream)
df_batch  = pd.read_csv(path_batch)

# === Pengaturan tampilan Pandas ===
pd.set_option('display.max_colwidth', 40)    
pd.set_option('display.colheader_justify', 'center')  
pd.set_option('display.max_columns', None)  
pd.set_option('display.width', 0)            

print("\n=== Dataset Streaming (5 Baris Pertama) ===")
display(df_stream.head(5))

print("\n=== Dataset Batch (5 Baris Pertama) ===")
display(df_batch.head(5))
```

```python
df_combined = pd.concat([df_batch, df_stream], ignore_index=True)

print("\n=== Dataset Gabungan (5 Baris Pertama) ===")
display(df_combined.head(5))

print("\n=== Info Dataset Gabungan ===")
df_combined.info()
```

```python
nan_counts = df_combined.isnull().sum()
print("=== Jumlah Data NaN per Kolom ===")
print(nan_counts)
print("\n" + "="*35)

duplicate_count = df_combined.duplicated(subset=['content']).sum()

print(f"\nJumlah data duplikat berdasarkan kolom 'content': {duplicate_count} baris")
print("="*35)
```

```python
print(f"Jumlah baris sebelum dibersihkan: {df_combined.shape[0]}")

df_cleaned = df_combined.drop_duplicates(subset=['content'], keep='first')
print(f"Jumlah baris setelah menghapus duplikat: {df_cleaned.shape[0]}")

df_cleaned = df_cleaned.dropna(subset=['content'])
print(f"Jumlah baris setelah menghapus NaN di 'content': {df_cleaned.shape[0]}")

columns_to_drop = ['userImage', 'reviewId', 'userName', 'replyContent', 'repliedAt', 'thumbsUpCount']
df_cleaned = df_cleaned.drop(columns=columns_to_drop)

print("\nKolom yang tersisa setelah pembersihan:")
print(df_cleaned.columns.tolist())

print("\n=== Jumlah Data NaN per Kolom (Setelah Pembersihan) ===")
print(df_cleaned.isnull().sum())
print("\n=== Dataset Bersih (5 Baris Pertama) ===")
display(df_cleaned.head())
```
