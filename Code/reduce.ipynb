```python
import pandas as pd
import numpy as np

file_input = "data_gabungan_gojek_cleaned.csv"
file_output = "data_gabungan_gojek_100k.csv"
jumlah_baris_target = 100000

try:
    df = pd.read_csv(file_input)
    print(f"File '{file_input}' berisi {len(df)} baris")

    print("Pengacakan urutan baris")
    df_shuffled = df.sample(frac=1, random_state=42).reset_index(drop=True)
    print(f"Pengacakan selesai")

    df_reduced = df_shuffled.head(jumlah_baris_target)

    df_reduced.to_csv(file_output, index=False)

    print(f"Berhasil diacak dan dikurangi menjadi {len(df_reduced)} baris")
    print(f"Disimpan ke '{file_output}'")

except FileNotFoundError:
    print(f"ERROR '{file_input}' tidak ditemukan")

except Exception as e:
    print(f"ERROR")
```
